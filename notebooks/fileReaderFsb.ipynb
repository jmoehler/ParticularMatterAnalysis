```python
from ibm_botocore.client import Config
import ibm_boto3

def upload_file_cos(credentials,local_file_name,key):  
    cos = ibm_boto3.client(service_name='s3',
    ibm_api_key_id=credentials['IBM_API_KEY_ID'],
    ibm_service_instance_id=credentials['IAM_SERVICE_ID'],
    ibm_auth_endpoint=credentials['IBM_AUTH_ENDPOINT'],
    config=Config(signature_version='oauth'),
    endpoint_url=credentials['ENDPOINT'])
    try:
        res=cos.upload_file(Filename=local_file_name, Bucket=credentials['BUCKET'],Key=key)
    except Exception as e:
        print(Exception, e)
    else:
        print(' File Uploaded')
```

## Der eigentliche Code
Da wie schon im Kapittel Partikulate Matter unter Partikulate Matter Projekt unter 1.Etappe unter Daten download erw채hnt wurde, wird in meinem Fall der Link nach einem einzigen Manenssystem erzeugt.
das macht man sich zu nutze und erzeugt einen allgemeinen Link, in welchen die Speziellen Daten dann automatisch eingef체gt werden.

```python
start = time.time()

now = datetime.datetime.now()


# es wird erst ein Loop mit jedem Jahr erzeugt
for y in [2017, 2018, 2019]:
    
    # in diesem einer mit jedem Monat
    for m in range(1,13):
        
        # in diesem einer mit jedem Tag
        for d in range(1,32):
            
            # die unterschiedlichen Monatsl채ngen werden Ber체cksichtigt
            if m in [4,6,9,11] and d > 30:
                continue
                
                # speziell der Februar
            if m == 2 and d >28:
                continue
                
            # es soll nicht nach zuk체nftigen Tagen gesucht werden
            if y == 2019 and m > now.month:
                continue 
                
            if y == 2019 and m == now.month and d > now.day -1:
                continue
            # der allg. Link in welchen alle diese Infos eingesetzt werden
            url = "http://archive.luftdaten.info/%04d-%02d-%02d/%04d-%02d-%02d_sds011_sensor_299.csv"  %(y,m,d,y,m,d)            
            # der neue Filename wird kreiert  
            filename = "%04d-%02d-%02d_sds011_sensor_299.csv"  %(y,m,d)
            localFile = "/var/tmp/%s"  %(filename)
            print("%s -> local:%s -> cos:%s" %(url, localFile, filename))
            try:
                urllib.request.urlretrieve(url, localFile)
                upload_file_cos(cgsCreds,localFile, filename)
            except:
                print("**** not available ****")
                continue
            

end = time.time()    
time = (end - start)
```
