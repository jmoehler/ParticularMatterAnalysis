```python
import pandas as pd
import numpy as np
from scipy import stats
from datetime import datetime
from sklearn import preprocessing
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

df = pd.read_csv("/Users/jakob/Dokumente/Schule/Jahresarbeit/Projekte/ParticulateMatterAnalysis/data/StuttgartCsvDone/Halbstd-Werte-Stuttgart-Mitte-SZ_2017_D.csv", index_col=[0], sep = ";")
print(df.head())



df.plot(figsize=(18,5))
```

```python
df1 = df.dropna()
print(df1)
```

```python
df1.columns
```

```python
df1 = df1[pd.to_numeric(df1["PM2.5 (µg/m³)-Mittel"], errors='coerce').notnull()]
```

```python
df1 = df1[pd.to_numeric(df1["PM10 (µg/m³)-Mittel"], errors='coerce').notnull()]
```

```python
df1.isnull().values.any()
```

```python
plt.scatter(df1["Temp. (°C)-Mittel"], df1["PM2.5 (µg/m³)-Mittel"])
```

```python
std_dev = 3
df2 = df1[(np.abs(stats.zscore(df1)) < float(std_dev)).all(axis=1)]
df2.plot(figsize=(18,5))
```

```python
df1.apply(lambda x:x.str.isalpha()).values.any()
```

```python
X = pd.DataFrame(df1["Temp. (°C)-Mittel"])
y = pd.DataFrame(df1["Feuchte (%)-Mittel"])
model = LinearRegression()
scores = []
kfold = KFold(n_splits=3, shuffle=True, random_state=42)
for i, (train, test) in enumerate(kfold.split(X, y)):
 model.fit(X.iloc[train,:], y.iloc[train,:])
 score = model.score(X.iloc[test,:], y.iloc[test,:])
 scores.append(score)
print(scores)
```
